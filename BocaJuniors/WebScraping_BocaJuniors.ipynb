Web Scraping o Scraping es un conjunto de técnicas que permiten extraer datos e información de una web, a través de su código HTML. La finalidad es poder recopilar grandes cantidades de datos cuyo uso posterior puede ser muy variado: homogenización de datos, tratamiento de contenido para la extracción de conocimiento, complementar datos en una web, etc.

```python
#Cargo lista con los nombres de los jugadores
jugadores = []

for i in list(range(0,len((jugadores_mas_titulos)),3)): 
  jugadores.append(jugadores_mas_titulos[i])
```

```python
#Cargo lista con los títulos totales de cada jugador
titulos_totales = []

for i in list(range(1,len((jugadores_mas_titulos)),3)): 
  titulos_totales.append(jugadores_mas_titulos[i])
```

```python
#Obtengo los títulos por cada jugador.
#Cargo listas para posteriormente, crear un diccionario para la generación del Dataframe.
#Subcadenas: defino las subcadenas con la palabra clave, lo que permitirá buscar cada título y obtener para cada uno de ellos el valor numérico que representa
#la cantidad de veces que se obtuvo dicho título.

#Inicialización de listas y generación de subcadenas.
lista_copaIntercontinental = []
subcadena_copaIntercontinental = 'Copa Intercontinental'
lista_copaLibertadores = []
subcadena_copaLibertadores = 'Copa Libertadores'
lista_copaSudamericana= []
subcadena_copaSudamericana = 'Copa Sudamericana'
lista_recopaSudamericana=[]
subcadena_recopaSudamericana = 'Recopa Sudamericana'
lista_primeraDivision = []
subcadena_primeraDivision = 'Primera División'
lista_copaArgentina = []
subcadena_copaArgentina = 'Copa Argentina'
lista_supercopaArgentina = []
subcadena_supercopaArgentina='Supercopa Argentina'
lista_copaLigaProfesional = []
subcadena_copaLigaProfesional='Copa de la Liga Profesional'
lista_carlos_ibarguren = []
subcadena_copa_carlos_ibarguren='Copa Carlos Ibarguren'
lista_copa_confraternidad=[]
subcadena_copa_confraternidad='Copa de Confraternidad Escobar-Gerona'
lista_copaBritanica = []
subcadena_britanica = 'Copa de Competencia Británica'


for i in list(range(2,len(jugadores_mas_titulos),3)):
  #Copa Argentina
  if jugadores_mas_titulos[i].find(subcadena_copaArgentina) !=-1:
    aux = []
    aux.append(jugadores_mas_titulos[i])
    aux = jugadores_mas_titulos[i].split('Copa Argentina')[0].strip()
    aux = aux[-1]
    lista_copaArgentina.append(aux)
  else:
    lista_copaArgentina.append(0)
  #Copa Británica
  if jugadores_mas_titulos[i].find(subcadena_britanica) !=-1:
    aux_britanica = []
    aux_britanica.append(jugadores_mas_titulos[i])
    aux_britanica = jugadores_mas_titulos[i].split('Copa de Competencia Británica')[0].strip()
    aux_britanica = aux_britanica[-1]
    lista_copaBritanica.append(aux)
  else:
    lista_copaBritanica.append(0)
  #Copa Intercontinental
  if jugadores_mas_titulos[i].find(subcadena_copaIntercontinental) !=-1:
    aux_intercontinental = []
    aux_intercontinental.append(jugadores_mas_titulos[i])
    aux_intercontinental = jugadores_mas_titulos[i].split('Copa Intercontinental')[0].strip()
    aux_intercontinental = aux_intercontinental[-1]
    lista_copaIntercontinental.append(aux_intercontinental)
  else:
    lista_copaIntercontinental.append(0)
  #Recopa Sudamericana
  if jugadores_mas_titulos[i].find(subcadena_recopaSudamericana) !=-1:
    aux_recopaSudamericana = []
    aux_recopaSudamericana.append(jugadores_mas_titulos[i])
    aux_recopaSudamericana = jugadores_mas_titulos[i].split('Recopa Sudamericana')[0].strip()
    aux_recopaSudamericana = aux_recopaSudamericana[len(aux_recopaSudamericana)-1]
    lista_recopaSudamericana.append(aux_recopaSudamericana)
  else:
    lista_recopaSudamericana.append(0)
  #Copa Libertadores
  if jugadores_mas_titulos[i].find(subcadena_copaLibertadores) !=-1:
    aux_copaLibertadores = []
    aux_copaLibertadores.append(jugadores_mas_titulos[i])
    aux_copaLibertadores = jugadores_mas_titulos[i].split('Copa Libertadores')[0].strip()
    aux_copaLibertadores = aux_copaLibertadores[len(aux_copaLibertadores)-1]
    lista_copaLibertadores.append(aux_copaLibertadores)
  else:
    lista_copaLibertadores.append(0)
  #Copa Sudamericana
  if jugadores_mas_titulos[i].find(subcadena_copaSudamericana) !=-1:
    aux_copaSudamericana = []
    aux_copaSudamericana.append(jugadores_mas_titulos[i])
    aux_copaSudamericana = jugadores_mas_titulos[i].split('Copa Sudamericana')[0].strip()
    aux_copaSudamericana = aux_copaSudamericana[len(aux_copaSudamericana)-1]
    lista_copaSudamericana.append(aux_copaSudamericana)
  else:
    lista_copaSudamericana.append(0)
   #Primera División
  if jugadores_mas_titulos[i].find(subcadena_primeraDivision) !=-1:
    aux_primeraDivision = []
    aux_primeraDivision.append(jugadores_mas_titulos[i])
    aux_primeraDivision = jugadores_mas_titulos[i].split('Primera División')[0].strip()
    aux_primeraDivision = aux_primeraDivision[len(aux_primeraDivision)-1]
    lista_primeraDivision.append(aux_primeraDivision)
  else:
    lista_primeraDivision.append(0)
  #Supercopa Argentina
  if jugadores_mas_titulos[i].find(subcadena_supercopaArgentina) !=-1:
    aux_supercopaArgentina = []
    aux_supercopaArgentina.append(jugadores_mas_titulos[i])
    aux_supercopaArgentina = jugadores_mas_titulos[i].split('Supercopa Argentina')[0].strip()
    aux_supercopaArgentina = aux_supercopaArgentina[len(aux_supercopaArgentina)-1]
    lista_supercopaArgentina.append(aux_supercopaArgentina)
  else:
    lista_supercopaArgentina.append(0)
  #Copa de la Liga Profesional
  if jugadores_mas_titulos[i].find(subcadena_copaLigaProfesional) !=-1:
    aux_copaLigaProfesional = []
    aux_copaLigaProfesional.append(jugadores_mas_titulos[i])
    aux_copaLigaProfesional = jugadores_mas_titulos[i].split('Copa de la Liga Profesional')[0].strip()
    aux_copaLigaProfesional = aux_copaLigaProfesional [len(aux_copaLigaProfesional )-1]
    lista_copaLigaProfesional.append(aux_supercopaArgentina)
  else:
    lista_copaLigaProfesional.append(0)
  #Copa Carlos Ibarguren
  if jugadores_mas_titulos[i].find(subcadena_copa_carlos_ibarguren) !=-1:
    aux_copa_carlos_ibarguren = []
    aux_copa_carlos_ibarguren.append(jugadores_mas_titulos[i])
    aux_copa_carlos_ibarguren = jugadores_mas_titulos[i].split('Copa Carlos Ibarguren')[0].strip()
    aux_copa_carlos_ibarguren = aux_copa_carlos_ibarguren [len(aux_copa_carlos_ibarguren )-1]
    lista_carlos_ibarguren.append(aux_copa_carlos_ibarguren)
  else:
    lista_carlos_ibarguren.append(0)
  #Copa Confraternidad
  if jugadores_mas_titulos[i].find(subcadena_copa_confraternidad) !=-1:
    aux_copa_confraternidad = []
    aux_copa_confraternidad.append(jugadores_mas_titulos[i])
    aux_copa_confraternidad = jugadores_mas_titulos[i].split('Copa de Confraternidad Escobar-Gerona')[0].strip()
    aux_copa_confraternidad = aux_copa_confraternidad [len(aux_copa_confraternidad )-1]
    lista_copa_confraternidad.append(aux_copa_confraternidad)
  else:
    lista_copa_confraternidad.append(0)
```
